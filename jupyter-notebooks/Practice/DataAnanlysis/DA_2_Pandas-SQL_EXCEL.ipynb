```python
sql = '''SELECT

    productname 'Product Name',

    categoryname 'Product Category',

    companyname 'Supplier Name'

FROM categories

INNER JOIN products

ON categories.categoryid = products.categoryid

INNER JOIN suppliers

ON products.supplierid = suppliers.supplierid

'''



df = pd.read_sql(sql, engine)

df
```

```python
import pandas as pd

def nan(value):
    if value =="NaN":
        return 0.15
    return value

def empty(value):
    if value =="":
        return 0.15
    return value

def notav(value):
    if value =="Not Available":
        return "Afganistan"
    return value

df = pd.read_excel('world-happiness-report.xlsx', 'world-happiness-report', index_col=False,
                  converters={
                    'Generosity':  nan,
                    'Social support':empty,
                    'Country name': notav
                  })

df
```

```python
import pandas as pd
df.to_excel('mydoc.xlsx', sheet_name = 'xyz')
```
