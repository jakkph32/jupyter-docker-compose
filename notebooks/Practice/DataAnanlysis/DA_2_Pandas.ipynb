```python
import pandas as pd

a = [3,4,6,7,9]
b = [1,3,6,3,4]
a_series = pd.Series(a)
b_series = pd.Series(b)
pd.concat([a_series,b_series], keys = ['First', 'Second'], names=['group, index'])
# keys for each dataframe and index can be any word but it define index of column
```

## Joins

```python
import pandas as pd
df_1 = pd.DataFrame({
    'Name':['Peter','Ahmed','James','Kate'],
    'Salary':[22000,35000,77000,25000]
})

df_2 = pd.DataFrame({
    'Name':['Peter','Ahmed','James','Kate'],
    'Department':['Sales','Admin','IT','HR']
})
df_3 = pd.merge(df_1, df_2, on='Name')
df_3
```

```python
import pandas as pd

df_1 = pd.DataFrame({
    'Name':['Peter','Ahmed','James','Kate'],
    'Salary':[22000,35000,77000,25000]
})
df_2 = pd.DataFrame({
    'Name':['Peter','Ahmed','Kate'],
    'Department':['Sales','Admin','HR']
})
df_3 = pd.merge(df_1, df_2, on='Name') # inner join
df_3
df_6 = pd.merge(df_1,df_2,on='Name', how='inner') # same as above # intersect
df_6
df_7 = pd.merge(df_1,df_2,on='Name', how='outer') #union
df_7
df_8 = pd.merge(df_1,df_2,on='Name', how='left') #all from left
df_8
df_9 = pd.merge(df_1,df_2,on='Name', how='right') #all from right
df_9
```

```python
import pandas as pd

df_1 = pd.DataFrame({
    'Name':['Peter','Ahmed','James','Kate'],
    'Salary':[22000,35000,77000,25000],
    'Department':['Sales','Admin','IT','HR'],
    'Education':['Bsc','Bsc','MSc','MA']
})
df_2 = pd.DataFrame({
    'Name':['Peter','Ahmed','James','Kate'],
    'Department':['Sales','Admin','IT','HR'],
    'Education':['Bsc','Bsc','MSc','MA']
})

df_3 = pd.merge(df_1,df_2,on='Name')
df_3 = pd.merge(df_1,df_2,on='Name', suffixes=('_first', '_second'))
df_3
```

```python
# checking NaN
import numpy as np
data= pd.read_csv('world-happiness-report.csv')
data
```

```python
pd.isnull(data) # Give true false
np.where(pd.isnull(data))
```

```python
data.iloc[25]
#data.iloc[25,4]
```

```python
good_data=data.dropna()
good_data
```

```python
bad_data=data[~data.index.isin(good_data.index)]
bad_data
```

```python
# droping nan data

data.dropna(inplace=True)
data
```

```python
# replacing all missing data
import pandas as pd

data= pd.read_csv('world-happiness-report.csv')
data.fillna(6000,inplace=True)
data[data.index.isin(bad_data.index)] # check 6000
```

```python
# replace specific column
import pandas as pd
data= pd.read_csv('world-happiness-report.csv')
data["Generosity"].fillna(7777,inplace=True)
good_data=data.dropna()
bad_data_generosity = data[~data[["Generosity"]].index.isin(good_data.index)]
bad_data_generosity
```

```python
# replace for mean, median, mode
mean= data["Social support"].mean()
data["Social support"].fillna(mean,inplace=True)
data.iloc[25]
```

```python
# Generate Random data
import numpy as np
import random
import pandas as pd
years=np.arange(1980,2020)
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
products=np.arange(1,20)
price=np.arange(1,1000)
values=[]
i=0

while i<1000:
    y=random.choice(years)
    m=random.choice(months)
    p=random.choice(products)
    prod=f'Product{p}'
    pr=random.choice(price)
    row=(y,m,prod,pr)
    values.append(row)
        
    
    i+=1

df=pd.DataFrame(values)
df.to_csv('Testdata.csv')
    
```
