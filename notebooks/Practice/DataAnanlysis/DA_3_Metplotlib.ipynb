```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

a = np.array([1,2,3,4,5,6])
b = np.array([4,5,6,7,3,4])
c = np.array([2,5,7,8,9,1])

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d') # 111 is size of plot
ax.scatter(a,b,c, s=70, alpha=0.5)
for d, e, f in zip(a, b, c):
    ax.plot3D([d, 5],[e, 5],[f, 5]) # here 5 is the starting point (make 0 for (0,0,0)
plt.show()
```

```python
# side by side bars
import numpy as np

import matplotlib.pyplot as plt

subjects = ('Java', 'SQL', 'Python', 'Neo4j', 'C++')
Peter = (100, 60, 15, 35, 38)
Jane  = (72, 85, 26, 67, 88)

bar_num = np.arange(5)
width = 0.25

plt.bar(bar_num, Peter, width, alpha = 0.5, color ='k', label='Peter')
plt.bar(bar_num +width, Jane, width, alpha = 0.5, color ='b', label='Jane')
plt.title('Performance Chart')
plt.xlabel('Peter and Jane')
plt.ylabel('Performance')
plt.xticks(bar_num + width, subjects)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#bar top and bottom
import numpy as np
import matplotlib.pyplot as plt
subjects = ['Java', 'SQL', 'Python', 'Neo4j', 'C++']
Peter = (100, 60, 15, 35, 38)
Jane  = (72, 85, 26, 67, 88)
#bar_num = np.arange(5)

plt.bar(subjects, Peter,  bottom = Jane, alpha = 0.5, color ='r', label='Peter')
plt.bar(subjects, Jane, alpha = 0.5, color ='b', label='Jane')
plt.title('Performance Chart')
plt.xlabel('Peter and Jane')
plt.ylabel('Performance')
plt.xticks(subjects)
plt.legend()
plt.tight_layout()
plt.show()
```

```python

# pie chart percentage and start angle 
import matplotlib.pyplot as plt
subjects = ['Java', 'SQL', 'Python', 'Neo4j', 'C++']
Peter = (100, 60, 15, 35, 38)
explode = (0, 0.1, 0, 0, 0) 
#plt.pie(Peter, explode=explode, labels=subjects, autopct='%1.1f%%', shadow=True, startangle=90)
fig, ax = plt.subplots()
ax.pie(Peter, explode=explode, labels=subjects, autopct='%1.1f%%', shadow=True, startangle=90)
ax.axis('equal') 
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt



products = ['Oil', 'Gold', 'Iron', 'Copper', 'Cocoa']
amount = [160, 30, 15, 12, 40]
countries = ['Brazil','Ukraine','Brazil','Ukraine','Brazil','Ukraine','Brazil','Ukraine','Brazil','Ukraine']
amountPercountry = [105, 55, 20, 10, 10, 5, 5, 7, 15, 25]
colors = ['#f0e184', '#cca300', 'black', '#a358a6', '#5c3b02']
country_colors = ['red', 'green','red', 'green','red', 'green','red','green','red', 'green']

plt.pie(amount, labels=products, colors=colors, startangle=90)

plt.pie(amountPercountry,labels=countries, colors=country_colors,radius=0.75,startangle=90, autopct='%1.1f%%')

inner_circle = plt.Circle((0,0),0.4,color='w',fc='white', linewidth=0)

f = plt.gcf()
f.gca().add_artist(inner_circle)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
